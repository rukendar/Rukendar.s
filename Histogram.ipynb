```python
import matplotlib.pyplot as plt
data = [5,6,7,7,8,9,10,10,10,11,12,12,12]
plt.hist(data,bins=7,edgecolor="black")
plt.title("Histogram")
plt.xlabel("value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(216)
data = np.random.randint(10,100,1000)
plt.hist(data,bins=10,edgecolor="black")
plt.title("Histogram of cancer patients with age distribution")
plt.xlabel("Age")
plt.ylabel("No.of.patients")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data ={
    "semester":["1st","2nd","1st","2nd","1st","2nd"],
    "hours_studied":[5,6,7,8,9,7]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8,6))
sns.barplot(x="semester",y="hours_studied",data=df)
plt.title("Hours studied per semester")
plt.xlabel("Semester")
plt.ylabel("Hours studied")
plt.show()
```
