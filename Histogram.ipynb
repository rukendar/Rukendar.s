```python
import matplotlib.pyplot as plt
data = [5,6,7,7,8,9,10,10,10,11,12,12,12]
plt.hist(data,bins=7,edgecolor="black")
plt.title("Histogram")
plt.xlabel("value")
plt.ylabel("Frequency")
plt.show()
```
